# Zomato data Analysis on Banglore Resturants

# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading csv file

```python
data=pd.read_csv('zomato.csv')
data.head()
```

```python
#This dataset gives avaliability of details in resturant
```

```python
data.shape
```

```python
data.columns
```

```python
df=data.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis=1)
df.head()
```

```python
#gives info of non-null values and datatypes
df.info()
```

# Dropping duplicate values

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['rate'].unique()
```

# Removing "NEW","-" and /5

```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
df['rate']=df['rate'].apply(handlerate)
df['rate'].head()
```

# Filling the null values in rate column with mean

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

# Dropping the null values

```python
df.dropna(inplace=True)
df.head()
```

```python
df.rename(columns={'approx_cost(for two people)':'Cost2plates','listed_in(type)':'Type'},inplace=True)
df.head()
```

```python
df['location'].unique()
```

As we see listed_in(city) and location both are almost same values. So lets keep one column

```python
df=df.drop(['listed_in(city)'],axis=1)
```

```python
df.head()
```

```python
df['Cost2plates'].unique() #removing comma because it is in string format.If its cost it has to be in float or int
```

# Removing, from Cost2plates column

```python
def handlecomma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
    
df['Cost2plates']=df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

```python
df.head()
```

# Cleaning rest_type of the Columns

```python
#checking the unique values in rest_type
rest_types=df['rest_type'].value_counts(ascending=False)
rest_types
```

```python
#values less than 1000 will be stored in rest_types_lessthan1000
rest_types_lessthan1000=rest_types[rest_types<1000]
rest_types_lessthan1000
```

# Making rest Types less than 1000 in frequency as other

```python
def handle_rests_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
    
df['rest_type']=df['rest_type'].apply(handle_rests_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
location=df['location'].value_counts(ascending=False)
location_lessthan300=location[location<300]

def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
    
df['location']=df['location'].apply(handle_location)
df['location'].value_counts()
```

# cleaning cuisines column

```python
cuisines = df['cuisines'].value_counts(ascending  = False)
cuisines_lessthan100 = cuisines[cuisines<100]



def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
